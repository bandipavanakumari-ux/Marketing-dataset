### 2.2 Visual Analysis

```python
len(Numerical_data)
```

**2.2.1 Histogram Shows data distribution.**

```python
plt.figure(figsize=(20,35), facecolor='skyblue')
plotnumber = 1

for column in Numerical_data:
    if plotnumber<=len(Numerical_data):
        ax = plt.subplot(8,3,plotnumber)
        plt.hist(df[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

**2.2.2 Box Plot**

```python
plt.figure(figsize=(20,35), facecolor='skyblue')
plotnumber = 1

for column in Numerical_data:
    if plotnumber<=len(Numerical_data):
        ax = plt.subplot(8,3,plotnumber)
        plt.boxplot(df[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

## 2) Categorical vs Numerical

```python
avg_income = df.groupby('Education')['Income'].mean()
avg_income 
```

```python
plt.bar(avg_income.index, avg_income.values)
plt.xticks(rotation=45)
plt.xlabel('Education')
plt.ylabel('Average Income')
plt.title('Average Income by Education')
plt.show()
```
